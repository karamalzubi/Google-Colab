```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns



cars = pd.read_csv('/content/CarPrice_Assignment (1).csv')

#cars.head() # returns the first 10 rows.
#cars.tail() # returns the last 5 rows
#cars.info() # gives the columns names, DataType and the nullable
#cars.describe() #
#cars.CarName.unique() # same as Select Distinct in SQL

companyname = cars['CarName'].apply(lambda x : x.split(' ')[0])

cars.insert(3,'companyname',companyname)

def replace_name(a,b):
  cars.companyname.replace(a,b,inplace = True)

replace_name ('maxda','mazda')
replace_name ('Nissan','nissan')
replace_name ('porcshce','porsche')
replace_name ('toyouta','toyota')
replace_name ('vokswagen','volkswagen')
replace_name ('vw','volkswagen')


#cars.companyname.unique()


cars['fueleconomy'] = (0.55 * cars['citympg']) + (0.45 * cars['highwaympg'])

#print(cars['fueleconomy'].to_string())

cars['price'] = cars['price'].astype('int')

sns.pairplot(cars)
plt.show()

#cars.to_string()

```
