4.Outliers Detection and Treatment : For columns: CarCount, BikeCount, BusCount, TruckCount, Total

Detect if any value is:

Less than 0 (which is impossible for counts).

Extremely high (using IQR method or quantile method).

If an outlier is found:

Replace it with the median of that column.

```python
import pandas as pd

# Step 1: Load the data
df = pd.read_csv('/content/traffic.csv')

# Step 2: Fill missing values using mean for counts
for col in ['CarCount', 'BikeCount', 'BusCount', 'TruckCount']:
    df[col].fillna(df[col].mean(), inplace=True)

df['Traffic Situation'].fillna(df['Traffic Situation'].mode()[0], inplace=True)

# Step 3: Create 'Total' column if it's not in the dataset
if 'Total' not in df.columns:
    df['Total'] = df['CarCount'] + df['BikeCount'] + df['BusCount'] + df['TruckCount']

# Step 4: Detect & Replace Outliers
def replace_outliers(column):
    Q1 = df[column].quantile(0.25)    # 25th percentile
    Q3 = df[column].quantile(0.75)    # 75th percentile
    IQR = Q3 - Q1                     # Interquartile range
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    median = df[column].median()

    # Replace any value < 0 or outside IQR bounds with the median
    df[column] = df[column].apply(lambda x: median if (x < 0 or x < lower or x > upper) else x)

# Apply to all required columns
for col in ['CarCount', 'BikeCount', 'BusCount', 'TruckCount', 'Total']:
    replace_outliers(col)

# Step 5: Print final cleaned data (optional, for confirmation)
print(df[['CarCount', 'BikeCount', 'BusCount', 'TruckCount', 'Total']].to_string())
```

5.Vehicle Distribution Analysis: Plot the distribution of CarCount, BikeCount, BusCount, and TruckCount:

Use Boxplots and Histograms for visualization.

Comment on which vehicle type shows the highest variability.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
data = pd.read_csv('/mnt/data/traffic.csv')

# Fill missing values with mean for numeric columns
for col in ['CarCount', 'BikeCount', 'BusCount', 'TruckCount']:
    data[col].fillna(data[col].mean(), inplace=True)

# Fill missing values for Traffic Situation with most frequent value
data['Traffic Situation'].fillna(data['Traffic Situation'].mode()[0], inplace=True)

# Create Total column if it doesn't exist
if 'Total' not in data.columns:
    data['Total'] = data['CarCount'] + data['BikeCount'] + data['BusCount'] + data['TruckCount']

# Function to handle outliers using IQR method
def handle_outliers(col_name):
    q1 = data[col_name].quantile(0.25)
    q3 = data[col_name].quantile(0.75)
    iqr = q3 - q1
    lower_limit = q1 - 1.5 * iqr
    upper_limit = q3 + 1.5 * iqr
    median_val = data[col_name].median()

    # Replace outliers and negatives with the median
    data[col_name] = data[col_name].apply(lambda val: median_val if (val < 0 or val < lower_limit or val > upper_limit) else val)

# Apply outlier handling
for column in ['CarCount', 'BikeCount', 'BusCount', 'TruckCount', 'Total']:
    handle_outliers(column)

# Plot histograms
plt.figure(figsize=(10, 6))
data[['CarCount', 'BikeCount', 'BusCount', 'TruckCount']].hist(bins=15, layout=(2, 2), color='skyblue', edgecolor='black')
plt.suptitle('Vehicle Count Distributions', fontsize=14)
plt.tight_layout()
plt.show()

# Plot boxplots
plt.figure(figsize=(8, 6))
sns.boxplot(data=data[['CarCount', 'BikeCount', 'BusCount', 'TruckCount']])
plt.title('Boxplot of Vehicle Counts')
plt.ylabel('Count')
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.show()
```

6.Traffic Situation Distribution: Visualize the counts of each Traffic Situation category (low, normal, high):

Use either a Bar Chart or Pie Chart.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('/mnt/data/traffic.csv')

# Fill missing values in 'Traffic Situation' with most frequent value
df['Traffic Situation'].fillna(df['Traffic Situation'].mode()[0], inplace=True)

# Count how many times each category appears
traffic_counts = df['Traffic Situation'].value_counts()

# Bar Chart
plt.figure(figsize=(6, 4))
traffic_counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Traffic Situation Distribution')
plt.xlabel('Traffic Level')
plt.ylabel('Number of Records')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

# Pie Chart (optional alternative)
plt.figure(figsize=(6, 6))
traffic_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['lightgreen', 'gold', 'salmon'])
plt.title('Traffic Situation Distribution (Pie Chart)')
plt.ylabel('')  # Hides y-axis label
plt.tight_layout()
plt.show()
```
